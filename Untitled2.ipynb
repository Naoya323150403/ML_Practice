```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from math import sqrt
from scipy.stats import skew
```

```python
train = pd.read_csv('train.csv',index_col='id')
test = pd.read_csv('test.csv',index_col='id')
submit = pd.read_csv('sample_submission.csv',index_col='id')
```

```python
train.head()
```

```python
target = train.pop('target')
X_train, X_test, y_train, y_test = train_test_split(train, target, train_size=0.70)
```

```python
dtrain = xgb.DMatrix(X_train, label=y_train)
```

```python
param = {
    'max_depth': 4, 
    'eta': 1, 
    'objective': 'reg:linear'
}
```

```python
model = xgb.train(param, dtrain)
```

```python
test.head()
```

```python
dtrain2 = xgb.DMatrix(test)
```

```python
predicted = model.predict(dtrain2)
```

```python
submit['target'] = predicted
```

```python
submit.head()
```

```python
submit.to_csv('submission3.csv')
```
